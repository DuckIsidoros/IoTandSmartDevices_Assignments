```python
import pandas as pd
```

```python
df = pd.read_csv('moisture_data.csv')

df.head(5)
```

```python
indices = [0, 40, 70, 100, 130, 160, 190]
readings = df.iloc[indices]['Moisture'].tolist()
```

```python
#@title Calculate the decrease for each 1 second
decreases = []
for i in range(1, len(readings)):
    diff = readings[i-1] - readings[i]
    decreases.append(diff)
```

```python
#@title # 4. Compute the final average
average_decrease = sum(decreases) / len(decreases)
```

```python
print("Result of this assignment:")
for i, val in enumerate(readings):
    label = "Dry" if i == 0 else f"{i}s Total"
    print(f"{label}: {val}")

print(f"\nIndividual Decreases: {decreases}")
print(f"Final Average Decrease per Second: {average_decrease:.2f}")
```
