```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('moisture.csv')

# Calculate GSM based on the instruction.
df['gsm'] = ((df['W_wet'] - df['W_dry']) / df['W_dry']) * 100


x = df['Cảm biến']
y = df['gsm']
m, c = np.polyfit(x, y, 1)

# Draw the calibration graph:
plt.scatter(x, y, color='blue', label='Dữ liệu thực tế')
plt.plot(x, m*x + c, color='red', label=f'Đường hiệu chuẩn: y = {m:.4f}x + {c:.4f}')
plt.xlabel('Giá trị Cảm biến')
plt.ylabel('Độ ẩm GSM (%)')
plt.title('Biểu đồ hiệu chuẩn cảm biến độ ẩm')
plt.legend()
plt.grid(True)
plt.show()
```
